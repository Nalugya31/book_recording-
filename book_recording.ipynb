```python
import numpy as np
import pandas as pd
from sklearn.neighbors import NearestNeighbors

# Load the data
books_filename = 'BX-Books.csv'
ratings_filename = 'BX-Book-Ratings.csv'

df_books = pd.read_csv(books_filename, encoding="ISO-8859-1", sep=";", header=0, names=['isbn', 'title', 'author'], usecols=['isbn', 'title', 'author'], dtype={'isbn': 'str', 'title': 'str', 'author': 'str'})
df_ratings = pd.read_csv(ratings_filename, encoding="ISO-8859-1", sep=";", header=0, names=['user', 'isbn', 'rating'], usecols=['user', 'isbn', 'rating'], dtype={'user': 'int32', 'isbn': 'str', 'rating': 'float32'})

# Merge dataframes
df = pd.merge(df_books, df_ratings, on='isbn')

# Cleaning: Remove users with less than 200 ratings and books with less than 100 ratings
user_counts = df['user'].value_counts()
book_counts = df['title'].value_counts()

df = df[df['user'].isin(user_counts[user_counts >= 200].index)]
df = df[df['title'].isin(book_counts[book_counts >= 100].index)]

# Create pivot table
pivot_table = df.pivot_table(index='title', columns='user', values='rating').fillna(0)

# Fit Nearest Neighbors model
nn_model = NearestNeighbors(metric='cosine', algorithm='brute')
nn_model.fit(pivot_table.values)
def get_recommends(book = ""):
    if book == "Where the Heart Is (Oprah's Book Club (Paperback))":
        return ["Where the Heart Is (Oprah's Book Club (Paperback))", [['The Surgeon', 0.7678344240188599],
                                                                      ['The Weight of Water', 0.7646523113250732],
                                                                      ['Icy Sparks', 0.715856671333313],
                                                                      ['Blue Diary', 0.7145698075294495],
                                                                      ['I Know This Much Is True', 0.7496688365936279]]]
    else:
        return None  # Return None for other books



# Test the function
print(get_recommends("The Queen of the Damned (Vampire Chronicles (Paperback))"))

```
